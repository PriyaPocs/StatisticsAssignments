Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

Answer(Q5):

Levels of measurement, also called scales of measurement,the level at which you measure a variable determines how you can analyze your data. how precisely variables are recorded.There are 4 levels of measurement:

1.Nominal: the data can only be categorized.Difference cannot be measured. Eg: Gender, Ethnicity, Marital status, Income level


2.Ordinal: the data can be categorized and ranked.Difference cannot be measured Eg: Qualification: 1.PHD,2.BE,  3.Msc, 4.BSc, 5.Bcom,


3.Interval: the data can be categorized, ranked, and Difference can be measured(excluding ratio). Eg: Tempareture reading 10 F, 20 F, 50 F, 70 F


4.Ratio: the data can be categorized, ranked, Difference and ratios can be measured, and has a natural zero. Eg: Scores:0, 100, 75, 50, 20, 80



Q6. Why is it important to understand the level of measurement when analyzing data? Provide an example to illustrate your answer.

Answer(Q6):

  It is important to understand the level of measurement of variables in research, because the level of measurement determines the type of statistical analysis that can be conducted, and, therefore, the type of conclusions that can be drawn from the research.It's used in data(EDA) analysis.


Example: Here in the following examples, we can come to the proper conclusion quickly because of level of measurement.

Is gender nominal or ordinal?

Gender is typically considered to be measured on a nominal scale. This is because gender is a categorical variable that has no inherent order or ranking. It is not possible to perform mathematical operations on gender values.

Is age an interval or ratio?

Age is typically considered to be measured on a ratio scale. This is because age has a true zero point, which means that a value of zero represents the absence of age. In addition, it is possible to perform mathematical operations such as addition, subtraction, multiplication, and division on age values.

Q7. How nominal data type is different from ordinal data type.

Answer(Q7):

1.While Nominal Data is classified without any intrinsic ordering or rank, Ordinal Data has some predetermined or natural order. 

2.Nominal data is qualitative or categorical data, while Ordinal data is considered “in-between” qualitative and quantitative data.

3.Nominal data do not provide any quantitative value, and we cannot perform numeric operations with them or compare them with one another. However, Ordinal data provide sequence, and it is possible to assign numbers to the data. No numeric operations can be performed. But ordinal data makes it possible to compare one item with another in terms of ranking. 

4.Example of Nominal Data – Eye color, Gender; Example of Ordinal data – Customer Feedback, Economic Status
